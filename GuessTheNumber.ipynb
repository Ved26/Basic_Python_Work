```python
import random
```

```python
import pandas as pd
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.high_score = 0
    
    def guess(self,x):
        self.x = x
        self.score = 100
        self.guessed_number = 0
        self.the_number = random.randint(1,self.x)
        print(f"\t\nInitial Score: {self.score}\n")
        while self.guessed_number != self.the_number and self.score>0:
            self.guessed_number = int(input(f"Guess the number between 1 and {self.x}: "))
            if self.guessed_number > self.the_number:
                self.score-=10
                print("\t\tSorry, Try Again! The number is too high")
                print(f"\t\tYour Current Score: {self.score}\n")
            elif self.guessed_number < self.the_number:
                self.score-=10
                print("\t\tSorry, Try Again! The number is too low")
                print(f"\t\tYour Current Score: {self.score}\n")

        if self.score==0:
            print("\n\tAww! You are out of guesses! Try Again next time! :)\n")
            print(f"\n\tThe answer was:{self.the_number}\n ")
            self.high_score = 0
        else:
            print(f"\n\tYay! You guessed the number {self.the_number}")
            print(f"\tYour Score: {self.score}\n")
            self.high_score = self.score
```

```python
def Create_Player(name):
    return Player(name)

def Difficulty():
    diff = {'Rookie': 50, 'Average': 100,'Pro': 250, 'Legend': 500}
    print("\n")
    print("\t\t1. Rookie(1-50)")
    print("\t\t2. Average(1-100)")
    print("\t\t3. Pro(1-250)")
    print("\t\t4. Legend(1-500)")
    val = input("\tChoose a Difficulty Level: ")
    try:
        level = int(val)
        if level > 0 and level < 5:
            return diff[list(diff.keys())[level-1]]
        else:
            raise ValueError()
    except:
        print("Invalid input. Please choose a number between 1 and 4.")
        return Difficulty()
        
print('\033[1m'+"\t\t\t\t\tHELLO! Welcome to the Guessing Game"+'\033[0;0m')
print('\033[1m'+"\t\t\t\t\t____________________________________\n"+'\033[0;0m')
print('\033[1m'+"\t\tRULES:"+'\033[0;0m')
print('\033[1m'+"\t\t\t1. Select a Difficulty Level\n"+'\033[0;0m')
print('\033[1m'+"\t\t\t2. I will generate a random number for you within the range 1 to your difficulty level.\n\t\t\t   If you select Rookie (say), the number will be within the range of 1-50"+'\033[0;0m')
print('\033[1m'+"\t\t\t3. Now, you will guess that random number within the range!\n "+'\033[0;0m')
print('\033[1m'+"\t\t\t4. You will Start the game with intial score of 100!\n "+'\033[0;0m')
print('\033[1m'+"\t\t\t\t. If the value you guessed is higher than the random number, you will be warned and"+'\033[0;0m')
print('\033[1m'+"\t\t\t\t  10 points will deducted from your score.\n"+'\033[0;0m')
print('\033[1m'+"\t\t\t\t. If the value you guessed is lower than the random number, you will be warned and"+'\033[0;0m')
print('\033[1m'+"\t\t\t\t  10 points will deducted from your score.\n"+'\033[0;0m')
print('\033[1m'+"\t\t\t\t. Guess the number and HURRAHH!. Your score will be displayed!"+'\033[0;0m')
print("\n\n")
print('\033[1m'+"\t\t\t\t\t\t\tLET'S PLAY"+'\033[0;0m')
print('\033[1m'+"========================================================================================================================="+'\033[0;0m')
game_on = True
while game_on:
    a = {}
    names = []
    high_scores = []
    i=1
    no_of_players = int(input("\tHow many Players will be playing? "))

    while i<=no_of_players:
        key = input("\t\tPlayer"+str(i)+" Name: ")
        names.append(key)
        value = Create_Player(key)
        a[key] = value
        i+=1

    turn = 1
    level = Difficulty()
    while turn <= no_of_players:
        print(f"{names[turn-1].upper()}'s Turn:")
        a[names[turn-1]].guess(level)
        high_scores.append(a[names[turn-1]].high_score)
        turn+=1

    print('\n\t')
    print("\033[1m"+"  LEADERBORD"+"\033[0;0m")
    leader_board = dict(zip(names,high_scores))
    lead = pd.DataFrame.from_dict(leader_board,orient='index',columns=['HighScore'])
    print(lead)
    if len(list(set(list(leader_board.values())))) == 1:
        print("\n\n It's a TIE!!")
    else:
        print("\n\n")
        print('\033[1m'+f'Congratulations! {max(leader_board, key=leader_board.get)}, You Won!!'+'\033[0;0m')
    
    ans = input("\n\nPlay Another Round? (y or n)")
    
    if ans != 'y':
        game_on = False
        
print('\033[1m'+'\n\t\tTHANK YOU FOR PLAYING :)'+'\033[0;0m')
```
