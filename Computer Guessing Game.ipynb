```python
import random 
```

```python
from time import sleep
```

```python
def Difficulty():
    diff = {'Rookie': 50, 'Average': 100,'Pro': 250, 'Legend': 500}
    loading = 'CHOOSING MY LEVEL!...'
    for i in range(len(loading)):
        print('\033[1m'+loading[i]+'\033[0;0m',end=' ', flush=True); sleep(0.1)
    level_value = random.choice(list(diff.values()))
    if level_value == 50:
        print("\n\n\tI am a "+'\033[1m'+'Rookie'+'\033[0;0m'+" Sir!"); sleep(0.2)
    elif level_value == 100:
        print("\n\n\tCome on! I am "+'\033[1m'+'Average'+'\033[0;0m' +" Though!"); sleep(0.2)
    elif level_value == 250:
        print("\n\n\tGame on Boy! I am a "+'\033[1m'+'Pro'+'\033[0;0m'+" here!"); sleep(0.2)
    elif level_value == 500:
        print("\n\n\tHuh! You sure about this? Challenging a "+'\033[1m'+'Legend?'+'\033[0;0m'); sleep(0.2)
    return level_value

def SmartGuess(x):
    low = 1
    high = x
    score = 100
    is_guessed = ''
    print(f'\n\tInitial Score: {score}')
    print(f'\tRange: 1-{high}')
    game_on = input("\nEnter Start: ").upper()
    if game_on=='Start'.upper():
        while is_guessed != 'CORRECT' and score>0:
            guessing = "GUESSING..."
            print("\n\t")
            for i in range(len(guessing)):
                print('\033[1m'+guessing[i]+'\033[0;0m',end=' ', flush=True); sleep(0.3)
            #my_guess = random.randint(low,high)
            my_guess = (low + high)//2
            print('\033[1m'+str(my_guess)+'\033[0;0m')
            feedback = input("\tToo High or Too Low? OR Is it Correct?  ").upper()

            if feedback == 'HIGH':
                high = my_guess-1
                score-=10
                print(f"\tScore: {score}")
            elif feedback == 'LOW':
                low = my_guess + 1
                score-=10
                print(f"\tScore: {score}")
            else:
                is_guessed = 'CORRECT'

        if(score==0):
            print('\033[1m'+"\n\t\tAh! I Give Up! I Lose :("+'\033[0;0m')
        else:
            print('\033[1m'+f"\n\n\t\tOh Yeahh!!! I guessed the Number!! It's {my_guess}. Come Tomorrow Kid!! :D"+'\033[0;0m')

    print('\033[1m'+'\n\tTHANK YOU FOR PLAYING :)'+'\033[0;0m')

    
def ComputerGuess(x):
    low = 1
    high = x
    if x > 100:
        SmartGuess(x)
        return
    
    score = 100
    is_guessed = ''
    print(f"\n\tInitial Score: {score}")
    print(f'\tRange: 1-{high}')
    game_on = input("\nEnter Start: ").upper()
    if game_on=='Start'.upper():
        while is_guessed != 'CORRECT' and score>0:
            guessing = "GUESSING..."
            print("\n\t")
            for i in range(len(guessing)):
                print('\033[1m'+guessing[i]+'\033[0;0m',end=' ', flush=True); sleep(0.3)
            my_guess = random.randint(low,high)
            #my_guess = (low + high)//2
            print('\033[1m'+str(my_guess)+'\033[0;0m')
            feedback = input("\tToo High or Too Low? OR Is it Correct?  ").upper()

            if feedback == 'HIGH':
                high = my_guess-1
                score-=10
                print(f"\tScore: {score}")
            elif feedback == 'LOW':
                low = my_guess + 1
                score-=10
                print(f"\tScore: {score}")
            else:
                is_guessed = 'CORRECT'
                print(f"\tScore: {score}")

        if(score==0):
            print('\033[1m'+"\n\t\tAh! I Give Up! I Lose :("+'\033[0;0m')
        else:
            print('\033[1m'+f"\n\n\t\tOh Yeahh!!! I guessed the Number!! It's {my_guess}. Come Tomorrow Kid!! :D"+'\033[0;0m')
    
    print('\033[1m'+'\n\tTHANK YOU FOR PLAYING :)'+'\033[0;0m')
```
